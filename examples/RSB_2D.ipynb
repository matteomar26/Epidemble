```julia
using Plots
```

```julia
N=35
L=20
r=2
X = Iterators.product(1:L, 1:L)
sx,sy = rand(1:L, N), rand(1:L, N)
R=zeros(Int,L,L)
for (x,y) in X
    for (xi,yi) in zip(sx,sy) 
        if (x-xi)^2 + (y-yi)^2 <= r^2
            R[x,y] = 1
        end
    end
end
```

```julia
scatter([x for (x,y) in X if R[x,y]==1], [y for (x,y) in X if R[x,y]==1], label="I")
scatter!([x for (x,y) in X if R[x,y]==0], [y for (x,y) in X if R[x,y]==0], label="S")
scatter!(sx,sy, grid=true, label="seed")
```

```julia
Q = ones(Int, L, L)
for (x,y) in X
    for (xi,yi) in X 
        if R[xi,yi] == 0 && (x-xi)^2 + (y-yi)^2 <= r^2
            Q[x,y] = 0
        end
    end
end
```

```julia

p1 = scatter([x for (x,y) in X if R[x,y]==1], [y for (x,y) in X if R[x,y]==1], label="candidate seed", 
        ticks=1:L, grid=:all, gridlw=1,color=:black, alpha=0.3,gridalpha=0.5, size=(500,500),showaxis=false, 
        legend=false, )
p1 = scatter!([x for (x,y) in X if Q[x,y]==1], [y for (x,y) in X if Q[x,y]==1], color=:red,label="true seed")
p1 = scatter!(sx,sy, label="seed!",color=:green)
```

```julia
N=100
L=20
r=2
X = Iterators.product(1:L, 1:L)
sx,sy = rand(1:L, N), rand(1:L, N)
R=zeros(Int,L,L)
for (x,y) in X
    for (xi,yi) in zip(sx,sy) 
        if (x-xi)^2 + (y-yi)^2 <= r^2
            R[x,y] = 1
        end
    end
end
Q = ones(Int, L, L)
for (x,y) in X
    for (xi,yi) in X 
        if R[xi,yi] == 0 && (x-xi)^2 + (y-yi)^2 <= r^2
            Q[x,y] = 0
        end
    end
end
p2 = scatter([x for (x,y) in X if R[x,y]==1], [y for (x,y) in X if R[x,y]==1], label="candidate seed", 
        ticks=1:L, grid=:all, gridlw=1,color=:black, alpha=0.3,gridalpha=0.5, size=(500,500),showaxis=false, 
        legend=false, )
p2 = scatter!([x for (x,y) in X if Q[x,y]==1], [y for (x,y) in X if Q[x,y]==1], color=:red,label="true seed")
p2 = scatter!(sx,sy, label="seed!",color=:green)
```

```julia
plot(p1,p2,size=(570,270))
```
